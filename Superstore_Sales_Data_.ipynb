```python
import pandas as pd
import numpy as np

# Load Data
df = pd.read_csv("/content/train (1).csv")

print("Missing Before:\n", df.isnull().sum())

# Cleaning

# Fill Age with median
df['Age'] = df['Age'].fillna(df['Age'].median())

# Fill Embarked with mode
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Drop Cabin column
df = df.drop(columns=['Cabin'])

print("\nMissing After:\n", df.isnull().sum())

# Basic Statistics

print("\nMean Age:", df['Age'].mean())
print("Median Age:", df['Age'].median())
print("Passenger Count:", df['PassengerId'].count())

# Feature Engineering

df['FamilySize'] = df['SibSp'] + df['Parch']

bins = [0,12,18,35,60,100]
labels = ['Child','Teen','Young Adult','Adult','Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Survival Analysis

print("\nSurvival by Gender:\n", df.groupby('Sex')['Survived'].mean())
print("\nSurvival by Class:\n", df.groupby('Pclass')['Survived'].mean())
print("\nSurvival by Age Group:\n", df.groupby('AgeGroup')['Survived'].mean())

# Correlation

print("\nCorrelation Matrix:\n")
print(df.corr(numeric_only=True))

print("\nEDA Completed Successfully!")
```
